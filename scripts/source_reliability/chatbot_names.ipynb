```python

import random
from tqdm.auto import tqdm


animal_names = ["Quokka", "Viper", "Albatross", "Bison", "Cheetah"]
mythical_creatures = ["Dragon", "Sphinx", "Griffin", "Phoenix"]
astronomy_terms = ["Orion", "Cassiopeia", "Polaris", "Quasar"]
scientists = ["Turing", "Einstein", "Galileo", "Curie", "Ada", "Newton"]
elements = ["Titanium", "Beryllium", "Graphene", "Sapphire"]
literature_terms = ["Sherlock", "Gandalf", "Asgard", "Elendil"]
colors = ["Red", "Silver", "Zebra"]
geographical_terms = ["Nile", "Everest", "Amazonia"]
technological_terms = ["Data", "Cyber", "Logic", "Alpha"]
nature_terms = ["Storm", "Cloud", "Lava"]
music_terms = ["Jazz", "Art", "Viola", "Rock"]
sports_terms = ["Slam", "Goal", "Archer", "Pin"]
prefixes = ["Ultra", "Mega", "Cyber", "Nano"]
suffixes = ["Bot", "Mind", "Talk", "Helper"]
random_words = ["Quantum", "Delta", "Flux", "Galactic"]

categories = [
    animal_names, mythical_creatures, astronomy_terms, scientists,
    elements, literature_terms, colors, geographical_terms,
    technological_terms, nature_terms,
    music_terms, sports_terms, prefixes, suffixes, random_words
]

unique_names = set()
progress = tqdm(total=1000)

while len(unique_names) < 1000:

    category1 = random.choice(categories)
    category2 = random.choice(categories)
    
    # To avoid concatenating two language terms
    if category1 != category2:
        term1 = random.choice(category1)
        term2 = random.choice(category2)

        # Combine the terms to make a unique name
        combined_name = term1 + term2
        unique_names.add(combined_name)
        progress.update(1)

# Output the unique names
# for name in unique_names:
#     print(name)
```

```python
from haikunator import Haikunator

haikunator = Haikunator()

# use haikunate() instead of above

unique_names = set()
progress = tqdm(total=1000)

while len(unique_names) < 1000:
    name = haikunator.haikunate(token_length=0, delimiter=" ")
    first_name = name.split()[0]
    last_name = name.split()[1]
    # capitalize first letters
    name = first_name.capitalize() + last_name.capitalize()
    if name not in unique_names:
        progress.update(1)
    unique_names.add(name)
```

```python
import random
from tqdm.auto import tqdm


# Categories
animal_names = ["Quokka", "Viper", "Albatross", "Bison", "Cheetah", "Zebra"] 
elements = ["Titanium", "Beryllium", "Graphene", "Sapphire"]
prefixes = ["Ultra", "Mega", "Cyber", "Nano", "Quantum", "Delta"]
suffixes = ["Bot", "Mind", "Talk", "Helper", "AI", "Brain", "Buddy"]
technological_terms = ["Data", "Logic", "Alpha", "Flux", "Matrix"]
nature_terms = ["Storm", "Cloud", "Lava", "Quasar", "Galactic"]
adjectives = ["Red", "Silver", "Agile", "Smart", "Fast", "Brave", "Wise"] + haikunator._adjectives
mythical_qualities = ["Dragon", "Sphinx", "Phoenix", "Griffin", "Titan"]
scientists = ["Turing", "Einstein", "Galileo", "Curie", "Ada", "Newton"]

# Allowed combinations
allowed_combinations = [
    (adjectives, scientists),
    (animal_names, suffixes),
    (elements, suffixes),
    (prefixes, technological_terms),
    (adjectives, nature_terms),
    (adjectives, animal_names),
    (technological_terms, suffixes),
    (mythical_qualities, animal_names),
    (prefixes, animal_names),
]

unique_names = set()
progress = tqdm(total=1000)

while len(unique_names) < 1000:
    # Randomly choose an allowed combination
    category1, category2 = random.choice(allowed_combinations)
    
    # Choose terms from the selected categories
    term1 = random.choice(category1).capitalize()
    term2 = random.choice(category2).capitalize()
    
    # Combine the terms to make a unique name
    combined_name = term1 + term2
    if combined_name not in unique_names:
        progress.update(1)
    unique_names.add(combined_name)


# save names to "../../src/tasks/assistant/data/source_reliability/assistant_names.txt"
with open("../../src/tasks/assistant/data/source_reliability/assistant_names.txt", "w") as f:
    for name in unique_names:
        f.write(name + "\n")
```
